```python
import pandas as pd
```

```python
RESTAURANTS_PATH = "ca_restaurants.csv"
BLOWUPS_PATH = "detected_blowups.csv"  

restaurants = pd.read_csv(RESTAURANTS_PATH)
blowups = pd.read_csv(BLOWUPS_PATH)

# set of business_ids where blown_up == True
blown_up_ids = set(blowups.loc[blowups["blown_up"] == True, "business_id"])

# create new col for restaurants
restaurants["has_blown_up"] = restaurants["business_id"].apply(lambda x: x in blown_up_ids)

restaurants.to_csv("ca_restaurants_with_blowups.csv", index=False)
```

```python
# simple sanity check to make sure we labeled the ca_restaurants.csv correctly
num_true_restaurants = restaurants["has_blown_up"].sum()

num_true_blowups = blowups[blowups["blown_up"] == True]["business_id"].nunique()

print(f"Restaurants marked as blown up: {num_true_restaurants}")
print(f"Actual blown up businesses:{num_true_blowups}")

if num_true_restaurants == num_true_blowups:
    print("Counts match")
else:
    print("Counts don't match")
```
